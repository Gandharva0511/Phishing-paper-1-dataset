```
#SVM classifier
import numpy as np
import matplotlib.pyplot as plt

# data to plot
n_groups = 6
means_frank = (90, 55, 40, 65)
means_guido = (85, 62, 54, 20)
p = (0.923148,.905954,.909958,.899444,.877701,.838395)
r = (0.827811,0.72699,0.704721,0.696936,.683407,.65383)
f1 = (0.872884,.860434,.865296,.865161,.862104,.844747)

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.25
opacity = 0.8

rects1 = plt.bar(index, p, bar_width,
alpha=opacity,
color='b',
label='Precision')

rects2 = plt.bar(index + bar_width, r, bar_width,
alpha=opacity,
color='g',
label='Recall')

rects3 = plt.bar(index + (2*bar_width), f1, bar_width,
alpha=opacity,
color='r',
label='F1 score')

plt.ylabel('Accuracy')
plt.xlabel('Phishing Vs Legitimate')
plt.title('SVM classifier')
plt.xticks(index + bar_width, ('1:1', '1:2', '1:3', '1:4','1:5','1:6'))
plt.legend()

plt.tight_layout()
plt.show()
```
